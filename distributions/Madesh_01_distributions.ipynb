#### 3.2 Categorical features - Level counts 

```python

fig, axs = plt.subplots(2, 2, figsize=(10, 7)) 
axs = axs.flatten() # converting from 2D to 1D array 
fig.suptitle("Categorical Features - Level Counts", fontsize=20, fontweight='bold')
colors = ['red', 'blue', 'green', 'yellow', "purple"]

#Plot 1 - 'room_type' 
rm_type_level_counts = data_df['room_type'].value_counts()
axs[0].set_title('Room_type', fontsize=16, fontweight='bold')
axs[0].bar(list(range(len(rm_type_level_counts))), rm_type_level_counts, tick_label=rm_type_level_counts.index, color=colors, edgecolor='black', linewidth=1.5, alpha=0.9)
axs[0].tick_params(axis='x', labelrotation=45, labelsize=10)
axs[0].set_ylabel('Listings', fontsize=10)

#Plot 2 - 'neighbourhood_group' 
neigh_group_level_counts = data_df['neighbourhood_group'].value_counts()
axs[1].set_title('Neighbourhood_Group', fontsize=16, fontweight='bold')
axs[1].bar(list(range(len(neigh_group_level_counts))), neigh_group_level_counts, tick_label=neigh_group_level_counts.index, color=colors, edgecolor='black', linewidth=1.5, alpha=0.9)
axs[1].tick_params(axis='x', labelrotation=45, labelsize=10)
axs[1].set_ylabel('Listings', fontsize=10)

#Plot 3 - 'neighbourhood' 
neigh_level_counts = data_df['neighbourhood'].value_counts()
axs[2].set_title('Neighbourhood', fontsize=16, fontweight='bold')
axs[2].fill_between(list(range(len(neigh_level_counts))), neigh_level_counts, color='red')
axs[2].set_xlabel('Neighborhood_number', fontsize=10)
axs[2].set_ylabel('Listings', fontsize=10)
axs[2].set_yscale('log')

#Plot 4 - 'host_name' 
host_name_counts = data_df['host_name'].value_counts()
axs[3].set_title('Host_name', fontsize=16, fontweight='bold')
axs[3].fill_between(list(range(len(host_name_counts))), host_name_counts, color='green')
axs[3].set_xlabel('Host_Name', fontsize=10)
axs[3].set_ylabel('Listings', fontsize=10)
axs[3].set_yscale('log')


plt.tight_layout()
plt.show()
```

#### 3.4 Numerical features - Distribution plots

##### 3.4.1 Latitude & Longitude- Distribution plots

```python
fig, axs=plt.subplots(1,3, figsize=(12,4))
axs=axs.flatten()

fig.suptitle('Listing location distributions', fontsize=20, fontweight='bold')

axs[0].set_title('Latitude', fontsize=16, fontweight='bold')
axs[0].hist(data_df['latitude'], color='red', bins=20)
axs[0].set_xlabel('Degrees')
axs[0].set_ylabel('Listings')

axs[1].set_title('Longitude',  fontsize=16, fontweight='bold')
axs[1].hist(data_df['longitude'], color='green', bins=20)
axs[1].set_xlabel('Degrees')
axs[1].set_ylabel('Listings')

axs[2].set_title('Location',  fontsize=16, fontweight='bold')
for neighborhood_group in data_df['neighbourhood_group'].unique():
    axs[2].scatter(
        data_df['longitude'][data_df['neighbourhood_group'] == neighborhood_group],
        data_df['latitude'][data_df['neighbourhood_group'] == neighborhood_group],
        s=0.2,
        label=neighborhood_group
    )
axs[2].set_xlabel('Longitude')
axs[2].set_ylabel('Latitude')
axs[2].legend(loc='upper left', markerscale=20)

plt.tight_layout()
plt.show()
```

The location plot confirms listings in a real map reference. As confirmed before, more listings in the Manhattan and Brooklyn areas compared to other neighbourhoods. 

##### 3.4.1 Other numerical features - Distribution plots

```python
numerical_features=['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
plot_titles=['Price', 'Minimum booking', 'Total reviews', 'Review rate', 'Listings per host', 'Vacancies last year']
xlabels=['USD', 'Nights', 'Reviews', 'Reviews per month', 'Listings per host', 'Avalible nights']
colors = ['red', 'blue', 'green', 'Orange', 'purple', 'violet']
fig, axs=plt.subplots(2,3, figsize=(10,6))
axs=axs.flatten()

fig.suptitle('Listing attribute distributions', fontsize=20, fontweight='bold')

for i, (plot_title, feature, xlabel, color) in enumerate(zip(plot_titles, numerical_features, xlabels, colors)):

    axs[i].set_title(plot_title)
    axs[i].hist(data_df[feature], bins=20, color=color)
    axs[i].set_xlabel(xlabel)
    axs[i].set_ylabel('Listings')
    axs[i].set_yscale('log')

plt.tight_layout()
```
